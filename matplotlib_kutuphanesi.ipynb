```python
import matplotlib.pyplot as plt

x=[1,3,5,7,9,11,13]
y=[40,50,60,70,80,90,100]

plt.plot(x,y)
plt.xlabel("X ekseni")
plt.ylabel("Y ekseni")
plt.title("Ã–rnek Grafik")
plt.bar(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)

plt.show()
```
